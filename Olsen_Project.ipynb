```python
# data on stocks using a data sheet from Kaggle
import plotly.graph_objects as go


import pandas as pd

# Create a Pandas DataFrame from a Google Sheet using the API
df = pd.read_csv('https://docs.google.com/spreadsheets/d/1ga_TZNhjn2VcOQKJ8LkL1zTo3FyfqxaKd7n2t25ZBcA/export?format=csv')

# drop column dividends and stock splits because they are not needed
columns_to_drop = ['Dividends', 'Stock Splits']

df = df.drop(columns=columns_to_drop)




# prompt: ask user what stock they would like to look up as an input. search Brand_Name or Ticker.
search_criteria = input("How would you like to search for the stock you are interested in? (Brand_Name, Ticker): ")
search_value = input(f"Enter the {search_criteria} for the stock you'd like to research: ")
if search_criteria == "Ticker":
    search_value = search_value.upper()

#filter data to only show the past 5 years
filtered_data = df[df[search_criteria] == search_value]
filtered_data = filtered_data[(filtered_data["Date"] >= "2019-01-01") & (filtered_data["Date"] <= "2024-12-31")]

figure = go.Figure(data=[go.Candlestick(x=filtered_data["Date"],
                                        open=filtered_data["Open"],
                                        high=filtered_data["High"],
                                        low=filtered_data["Low"],
                                        close=filtered_data["Close"])])

# Update the layout
figure.update_layout(xaxis_rangeslider_visible=False)

# Show the figure
figure.show()
```
